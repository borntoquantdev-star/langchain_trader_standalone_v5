# 📗 หน่วยที่ 2 — LLM กับการเทรด

```
main()                             ← หลัก
├── setup_llm()                    ← รอง
├── analyze_price_data()           ← รอง
├── analyze_with_system_prompt()   ← รอง
├── analyze_candlestick()          ← รอง
├── analyze_support_resistance()   ← รอง
└── show_limitations()             ← รอง
```

````python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage

load_dotenv()
OPENAI_API_KEY = os.getenv('OPENAI_API_KEY')
OPENAI_MODEL = os.getenv('OPENAI_MODEL', 'gpt-4o-mini')

PRICE_DATA = """XAUUSD (15m):
แท่ง 1: O=2648.50 H=2651.20 L=2647.80 C=2650.10
แท่ง 2: O=2650.10 H=2653.40 L=2649.50 C=2652.80
แท่ง 3: O=2652.80 H=2655.00 L=2651.90 C=2654.20
แท่ง 4: O=2654.20 H=2654.80 L=2650.30 C=2651.00
แท่ง 5: O=2651.00 H=2652.50 L=2648.00 C=2649.30"""

CANDLE_DATA = """XAUUSD 3 แท่งล่าสุด:
แท่ง 1 (Bear): O=2660 H=2661 L=2648 C=2650
แท่ง 2 (Doji): O=2650 H=2652 L=2649 C=2651
แท่ง 3 (Bull): O=2651 H=2665 L=2650 C=2663"""

SR_DATA = """XAUUSD 10 แท่ง (4H):
High: 2680, 2675, 2670, 2685, 2678, 2690, 2682, 2688, 2695, 2683
Low:  2655, 2660, 2650, 2665, 2658, 2670, 2662, 2668, 2672, 2665
ราคาปัจจุบัน: 2675"""

print(f'✅ Model: {OPENAI_MODEL}')
````

## 🔧 ฟังก์ชันรอง 1: `setup_llm()`

````python
def setup_llm():
    return ChatOpenAI(model=OPENAI_MODEL, api_key=OPENAI_API_KEY, temperature=0.1)

# เรียกใช้
llm = setup_llm()
print('✅ LLM พร้อม')
````

## 📊 ฟังก์ชันรอง 2: `analyze_price_data()`

ส่ง OHLC → AI บอกขาขึ้น/ลง/ทรงตัว

````python
def analyze_price_data(llm, data):
    print('📊 === วิเคราะห์ราคา ===')
    r = llm.invoke([
        SystemMessage(content='ดูราคาแล้วบอกขาขึ้น ลง หรือทรงตัว'),
        HumanMessage(content=f'{data}\n\nตลาดเป็นขาอะไร?'),
    ])
    print(f'🤖: {r.content}')

# เรียกใช้
analyze_price_data(llm, PRICE_DATA)
````

## 🎯 ฟังก์ชันรอง 3: `analyze_with_system_prompt()`

Trading System Prompt → AI ตอบเป็นโครงสร้าง Entry/SL/TP

````python
def analyze_with_system_prompt(llm, data):
    print('🎯 === Trading Prompt ===')
    r = llm.invoke([
        SystemMessage(content='นักวิเคราะห์ XAUUSD\nไม่ชัด→HOLD | R:R>=1:2\n'
            'ตอบ: แนวโน้ม / แนะนำ / Entry / SL / TP / เหตุผล'),
        HumanMessage(content=data),
    ])
    print(f'🤖:\n{r.content}')

# เรียกใช้
analyze_with_system_prompt(llm, PRICE_DATA)
````

## 🕯️ ฟังก์ชันรอง 4: `analyze_candlestick()`

ส่ง pattern แท่งเทียน → AI บอกชื่อ + Bullish/Bearish

````python
def analyze_candlestick(llm, data):
    print('🕯️ === Candlestick ===')
    r = llm.invoke([
        SystemMessage(content='Candlestick expert: 1)ชื่อ 2)Bull/Bear 3)ควรทำอะไร'),
        HumanMessage(content=data),
    ])
    print(f'🤖: {r.content}')

# เรียกใช้
analyze_candlestick(llm, CANDLE_DATA)
````

## 📐 ฟังก์ชันรอง 5: `analyze_support_resistance()`

ส่ง High/Low → AI หาแนวรับ/แนวต้าน

````python
def analyze_support_resistance(llm, data):
    print('📐 === S/R ===')
    r = llm.invoke([
        SystemMessage(content='หา Support/Resistance จาก High/Low'),
        HumanMessage(content=data),
    ])
    print(f'🤖: {r.content}')

# เรียกใช้
analyze_support_resistance(llm, SR_DATA)
````

## ⚠️ ฟังก์ชันรอง 6: `show_limitations()`

LLM ไม่มี real-time data → ต้องใช้ Tools (หน่วยที่ 3)

````python
def show_limitations(llm):
    print('⚠️ === ข้อจำกัด ===')
    print(f'🤖: {llm.invoke("ราคาทอง XAUUSD ตอนนี้?").content}')
    print('❌ ไม่มี real-time | ❌ ดึง MT5 ไม่ได้ | ❌ คำนวณ indicators ไม่ได้')
    print('💡 ทางออก → Tools & Agent!')

# เรียกใช้
show_limitations(llm)
````

---
## ★ ฟังก์ชันหลัก `main()` — รันทุกอย่างรวมกัน

````python
def main():
    llm = setup_llm()                              # รอง 1
    analyze_price_data(llm, PRICE_DATA)            # รอง 2
    analyze_with_system_prompt(llm, PRICE_DATA)    # รอง 3
    analyze_candlestick(llm, CANDLE_DATA)          # รอง 4
    analyze_support_resistance(llm, SR_DATA)       # รอง 5
    show_limitations(llm)                          # รอง 6
    print('\n✅ จบหน่วยที่ 2!')

# เรียกใช้
main()
````
