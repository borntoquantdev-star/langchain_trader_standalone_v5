````python
%pip install langchain langchain-openai python-dotenv -q

import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage

load_dotenv()
OPENAI_API_KEY = os.getenv('OPENAI_API_KEY', '')
OPENAI_MODEL = os.getenv('OPENAI_MODEL', 'gpt-4o-mini')
print(f'✅ Model: {OPENAI_MODEL}')
````

## 🔧 ฟังก์ชันรอง 1: `setup_llm()`

`ChatOpenAI` = object เชื่อมต่อกับ GPT ผ่าน API

````python
def setup_llm(temperature=0.7):
    """สร้าง ChatOpenAI"""
    llm = ChatOpenAI(model=OPENAI_MODEL, api_key=OPENAI_API_KEY, temperature=temperature)
    print(f'✅ LLM พร้อม (temp={temperature})')
    return llm

# เรียกใช้
llm = setup_llm()
````

## 🚀 ฟังก์ชันรอง 2: `demo_first_prompt()`

`llm.invoke("ข้อความ")` → ส่ง prompt → ได้คำตอบ

````python
def demo_first_prompt(llm):
    """ส่ง prompt แรก"""
    print('\n🚀 === ส่ง Prompt แรก ===')
    r = llm.invoke('สวัสดี! คุณคือใคร? ตอบสั้นๆ 2 บรรทัด')
    print(f'🤖: {r.content}')

# เรียกใช้
demo_first_prompt(llm)
````

## 💬 ฟังก์ชันรอง 3: `demo_messages()`

- `SystemMessage` → กำหนดบุคลิก AI
- `HumanMessage` → คำถามของเรา
- `AIMessage` → คำตอบเก่า (ส่งเป็น history)

````python
def demo_messages(llm):
    """เปลี่ยน SystemMessage → เปลี่ยนบุคลิก AI"""
    print('\n💬 === Messages ===')
    r1 = llm.invoke([
        SystemMessage(content='คุณเป็นครูสอนการเงิน ตอบสั้นๆ ใช้อิโมจิ'),
        HumanMessage(content='แนวรับ คืออะไร?'),
    ])
    print(f'👨‍🏫 ครู: {r1.content}')
    r2 = llm.invoke([
        SystemMessage(content='คุณเป็นเทรดเดอร์มืออาชีพ ใช้ศัพท์เทคนิค'),
        HumanMessage(content='แนวรับ คืออะไร?'),
    ])
    print(f'📊 Pro: {r2.content}')
    print('💡 คำถามเดิม คำตอบต่าง → เพราะ SystemMessage ต่างกัน!')

# เรียกใช้
demo_messages(llm)
````

## 🌡️ ฟังก์ชันรอง 4: `demo_temperature()`

- `temp=0` → ตอบเหมือนกันทุกครั้ง (วิเคราะห์)
- `temp=1` → ตอบต่างทุกครั้ง (creative)

````python
def demo_temperature():
    """เปรียบเทียบ temperature"""
    print('\n🌡️ === Temperature ===')
    q = 'ราคาทองจะขึ้นหรือลง? ตอบ 1 บรรทัด'
    cold = setup_llm(0.0)
    print(f'🧊 temp=0 (1): {cold.invoke(q).content}')
    print(f'🧊 temp=0 (2): {cold.invoke(q).content}')
    hot = setup_llm(1.0)
    print(f'🔥 temp=1 (1): {hot.invoke(q).content}')
    print(f'🔥 temp=1 (2): {hot.invoke(q).content}')

# เรียกใช้
demo_temperature()
````

## ✍️ ฟังก์ชันรอง 5: `demo_prompt_engineering()`

เทคนิค: กำหนดบทบาท + ระบุ format + ให้ข้อจำกัด

````python
def demo_prompt_engineering(llm):
    """prompt แย่ vs ดี"""
    print('\n✍️ === Prompt Engineering ===')
    print(f'❌ แย่: {llm.invoke("ทองคำ").content[:80]}...')
    good = llm.invoke('คุณเป็นนักวิเคราะห์ทองคำ สมมติราคา 2650 USD\n'
        'ตอบ: แนวโน้ม / เหตุผล / แนะนำ BUY/SELL/HOLD').content
    print(f'\n✅ ดี:\n{good}')

# เรียกใช้
demo_prompt_engineering(llm)
````

## 🔄 ฟังก์ชันรอง 6: `demo_chat_history()`

LLM ไม่มีความจำ → ส่ง history ไปด้วยถึงจะจำได้

````python
def demo_chat_history(llm):
    """AI จำได้ถ้าส่ง history"""
    print('\n🔄 === Chat History ===')
    print(f'❌ ไม่มี history: {llm.invoke("ฉันชื่ออะไร?").content}')
    r = llm.invoke([
        HumanMessage(content='ชื่อฉันคือ สมชาย'),
        AIMessage(content='สวัสดีครับคุณสมชาย!'),
        HumanMessage(content='ฉันชื่ออะไร?'),
    ])
    print(f'✅ มี history: {r.content}')

# เรียกใช้
demo_chat_history(llm)
````

---
## ★ ฟังก์ชันหลัก `main()` — รันทุกอย่างรวมกัน

````python
def main():
    llm = setup_llm()                    # รอง 1
    demo_first_prompt(llm)               # รอง 2
    demo_messages(llm)                   # รอง 3
    demo_temperature()                   # รอง 4
    demo_prompt_engineering(llm)         # รอง 5
    demo_chat_history(llm)               # รอง 6
    print('\n✅ จบหน่วยที่ 1!')

# เรียกใช้
main()
````
